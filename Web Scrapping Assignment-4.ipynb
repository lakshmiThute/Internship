Q.6) Scrape the details of top 100 songs on billiboard.com.
Url = https:/www.billboard.com/
You have to find the following details:
A) Song name
B) Artist name
C) Last week rank
D) Peak rank
E) Weeks on board

```python
# Opening the web driver.

driver=webdriver.Chrome(r'C:\Users\91749\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
# Getting the URL.

url = 'https://www.billboard.com/charts/hot-100/'
driver.get(url)
```

```python
# Creating empty lists.

Song_Name =[]
Singer=[]
rank=[]
Last_Week=[]
Weeks_on_board=[]
```

```python
# Scrapping the song element.

song=driver.find_elements("xpath","//li[@Class='lrv-u-width-100p']")
for i in song:
    if i.text is None :
        Song_Name.append("--") 
        Singer.append("--") 
        rank.append("--") 
        Last_Week.append("--") 
        Weeks_on_board.append("--") 
    else:
        Song_Name.append(i.text.split('\n')[0])
        Singer.append(i.text.split('\n')[1])
        rank.append(i.text.split('\n')[2])
        Last_Week.append(i.text.split('\n')[3])
        Weeks_on_board.append(i.text.split('\n')[4])
```

```python
# Creating Dataframe.

Billboard=pd.DataFrame()
Billboard['Song_Name']=Song_Name
Billboard['Peak_rank']=rank
Billboard['Singer / Crew']=Singer
Billboard['Last_Week_Rank']=Last_Week
Billboard['Weeks_on_board']=Weeks_on_board
Billboard
```
