```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://cricangel.com/odi-womens-all-rounder/')
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
team=soup.find_all('td',class_="column-2")
team
```

```python
name=[]
for player in team:
    name.append(player.get_text().strip())
name    
```

```python
logo=soup.find_all('td',class_="column-3")
logo
```

```python
nation=[]
for odi in logo:
    nation.append(odi.get_text().strip())
nation    
```

```python
rate=soup.find_all('td',class_="column-4")
rate
```

```python
rates=[]
for rating in rate:
    rates.append(rating.get_text().strip())
rates    
```

```python
import pandas as pd
df=pd.DataFrame({'Player Name':name,'Team Name':nation,'Rating':rates})
df.head(10)
```
