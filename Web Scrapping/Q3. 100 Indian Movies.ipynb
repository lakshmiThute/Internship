```python
# Q3. Write a python program to display IMDB’s Top rated 100 Indian movies’ data.
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.imdb.com/list/ls009997493/?sort=user_rating,desc&st_dt=&mode=simple&page=1&ref_=ttls_vw_smp')
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
top_movies=soup.find_all('div',class_="col-title")
top_movies
```

```python
movies=[]
for movie in top_movies:
    movies.append(movie.get_text().replace('\n',''))
movies    
```

```python
top_rated=soup.find_all('div',class_="col-imdb-rating")
top_rated
```

```python
ratings=[]
for rating in top_rated:
    rating=rating.get_text().strip()
    ratings.append(rating)
ratings    
```

```python
import pandas as pd
data=pd.DataFrame()
data['Movies Names']=movies
data['Ratings']=ratings
data.head()
```
