```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.icc-cricket.com/rankings/womens/player-rankings/odi')
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
batting=soup.find_all('td',class_="table-body__cell name")
batting
```

```python
women=[]
for womens in batting:
    women.append(womens.get_text().strip())
women    
```

```python
team=soup.find_all('span',class_="table-body__logo-text")
team
```

```python
Teams=[]
for player in team:
    Teams.append(player.get_text().strip())
Teams    
```

```python
rating=soup.find_all('td',class_="table-body__cell u-text-right rating")
rating
```

```python
rates=[]
for rate in rating:
    rates.append(rate.get_text().strip())
rates    
    
```

```python
import pandas as pd
df=pd.DataFrame()
df['Player Name']=women
df['Team Name']=Teams
df['Rating']=rating
df.head()
```
